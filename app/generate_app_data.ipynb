```python
import pandas as pd
```

```python
data_path = "../notebooks/translations"
df_sent_ee = pd.read_parquet(f"{data_path}/ee_sentences.parquet")
```

```python
df_sent_grc = pd.read_parquet(f"{data_path}/grc_sentences.parquet")
```

```python
df_words_grc = pd.read_parquet(f"{data_path}/grc_words.parquet")
```

```python
df_sent_grc_filtered = df_sent_grc.loc[df_sent_ee.grc_sent_idx]
```

```python
df_words_grc_filtered = df_words_grc.loc[df_sent_ee.grc_target_idx]
```

```python
df_words_grc_filtered
```

```python
df_sent_grc_filtered
```

```python
df_app_data_merged = pd.merge(df_sent_ee, df_sent_grc_filtered[["metadata", "sentence_obj", "sentence_txt"]], left_on="grc_sent_idx", right_index=True)
df_app_data_merged = pd.merge(df_app_data_merged, df_words_grc_filtered[["index", "text", "lemma"]], left_on="grc_target_idx", right_index=True)
df_app_data_merged = df_app_data_merged.rename(columns={
    "index": "target_grc_repr",
    "text": "target_grc"
})
```

```python
def replace_with_mask(row):
    return [w | {"masked": w["string"] == row["target_grc"]} for w in row["sentence_obj"] ]
```

```python
#df_app_data_merged["sentence_obj_masked"] = 
df_app_data_merged["sentence_obj_masked"] = df_app_data_merged.apply(replace_with_mask, axis=1)
```

```python
df_app_data_merged["memory_score"] = 0
```

```python
df_app_data_merged.iloc[0]["sentence_obj_masked"]
```

```python
df_app_data_merged.to_parquet("app_data.parquet")
```
